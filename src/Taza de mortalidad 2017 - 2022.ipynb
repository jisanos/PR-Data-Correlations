La data original contiene alrededor de 90k entradas. Para reducir la data y trabajar con una muestra mas pequeña, se utilizó la función dropna() para eliminar todas las entradas que contengan una columna con valores NaN. También la data original contiene varias columnas que no nos interesa tener, por lo que solo se muestran aquellas que son pertinentes para el análisis. 

```python
data = pd.read_csv('/Users/angelromero/Documents/Este semestre/CCOM4995/Proyecto final/Data Original/mortalidad-2017-2019.csv', encoding='latin-1')
data

# Get only the columns that we want to analize
df = pd.DataFrame(data, columns=["Gender", 
                                 "Age", 
                                 "MaritalStatus", 
                                 "DeathPlace", 
                                 "MunicipalityDeathPlace", 
                                 "DeathDate_Year", 
                                 "DeathCause_I (Desription)",
                                 "DeathCause_II (Desription)",
                                 "DeathCause_III (Desription)",
                                 "TypeOfDeath",
                                 "Education",
                                 "ACMEUnderlyingCauseDescription"])
# Rename some columns
df.rename(columns={"DeathCause_I (Desription)":"Ailments I", 
                   "DeathCause_II (Desription)":"Ailments II",
                   "DeathCause_III (Desription)":"Ailments II",
                   "ACMEUnderlyingCauseDescription": "FinalDeathCause"}, inplace = True)


df.reset_index(drop=True,inplace=True)
df
# df.dtypes
```

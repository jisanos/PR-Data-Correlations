```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../data/raw/po-total-pr.csv", encoding="utf-16", delimiter='\t')
df
```

```python
# Rename columns
df = df.rename(columns={
    "Municipio": "Municipality",
    "Latitude (generated)": "Latitude",
    "Longitude (generated)": "Longitude",
    "Cambio desde abril/2020  (#)": "Change since April 2020 (#)",
    "Cambio desde abril/2020  (%)": "Change since April 2020 (%)",
    "Pob. Total": "Total Population",
})
# Remove useless columns (e.g. "Geometry")
df = df.drop("Geometry", axis=1) if "Geometry" in df else df
df
```

```python
# Sort values by biggest municipality first.
df.sort_values(by='Total Population', ascending=False)
```

```python
# Output the DataFrame to CSV, and overwrite if the file already exists.
df.to_csv("../../data/clean/po-total-pr-clean.csv", index=False)
```

# Municipality Population Change Totals
- https://censo.estadisticas.pr/EstimadosPoblacionales

```python
import numpy as np
import pandas as pd
```

```python
# Import the pop change totals Excel and clean useless formatting rows.
pob_change_totals_mun = pd.read_excel("../../data/raw/po-pr-change-totals-municipalities.xlsx", sheet_name="PRM-EST2022-CHG", skiprows=5)
pob_change_totals_mun = pob_change_totals_mun[:-7]
# Drop any derivative columns (this is the core data)
pob_change_totals_mun = pob_change_totals_mun.iloc[:, :5]
pob_change_totals_mun
```

```python
# Fix the municipality names.
for index, row in pob_change_totals_mun.iterrows():
    pob_change_totals_mun.at[index, "Puerto Rico"] = row["Puerto Rico"].replace(".", "").replace("Municipio, Puerto Rico", "")
pob_change_totals_mun
```

```python
# Clean up column names.
pob_change_totals_mun.columns = ["State", "Estimates (2020)", "2020", "2021", "2022"]
pob_change_totals_mun
```

```python
pob_change_totals_mun.reset_index(drop=True, inplace=True)
pob_change_totals_mun.sort_values(by=["2022"], ascending=False, inplace=True)
pob_change_totals_mun
```

# Final Results

```python
pob_change_totals_mun.to_csv("../../data/clean/po-pr-total-change-clean.csv", index=False)
```
